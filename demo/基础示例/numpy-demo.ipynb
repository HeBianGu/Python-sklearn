## 生成随机矩阵

```python
np.random.random(size=(20,30))#随机生成20*30的矩阵
```

```python
np.random.normal(0,1,size=(16,5))#随机生成均值为0方差为1的正态分布，其实就是标准正态分布
```

## 生成服从标准分布的矩阵

```python
np.random.normal(size=(16,5))#生成标准正态分布
```

## 再某个区间生成随机数组

```python
np.random.uniform(0.0,100,size=10)
```

## 随机数组生成（本质就是size传一维）

```python
r=np.random.normal(0,50,size=10)#正态分布
print(r)
r=np.random.randint(0,50,size=10)#随机分布
print(r)
```

## 随机矩阵生成（本质就是size传多维）

```python
r=np.random.normal(0,50,size=(2,3))#正态分布
print(r)
r=np.random.randint(0,50,size=(2,3))#随机分布
print(r)
```
