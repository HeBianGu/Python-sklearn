## 绘制散点图

```python
x=np.random.randint(100,size=50)
```

```python
y=np.random.randint(100,size=50)
```

```python
plt.scatter(x,y)
plt.show()
```

## 绘制多颜色值散点图

```python
for x_i,y_i in zip(x,y):
    if x_i==y_i:
        plt.scatter(x_i,y_i,color='red')
    if x_i>y_i:
        plt.scatter(x_i,y_i,color='green')
    if x_i<y_i:
        plt.scatter(x_i,y_i,color='blue')
plt.show()
```

### 绘制三维

```python
import numpy as np
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.hot)
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.cm.hot)
ax.set_zlim(-2,2)
show()
```

## 绘制等值线

```python
from matplotlib.pyplot import *
def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)
n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)
contourf(X, Y, f(X,Y), 8, alpha=.75, cmap=cm.hot)
C = contour(X, Y, f(X,Y), 8, colors='black', linewidth=0.5)
clabel(C, inline=1, fontsize=10)#绘制值
show()
```

## 平面网格视图

```python
from matplotlib.pyplot import *
def f(x,y):
     return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)
n = 10
x = np.linspace(-3,3,4*n)
y = np.linspace(-3,3,3*n)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
imshow(Z,interpolation='nearest', cmap='bone', origin='lower')
colorbar(shrink=0.92)
show()
```

## 测试

```python
from matplotlib.pyplot import *
def f(x,y):
    return (1-x)**2 +2*(y-x**2)**2

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-1,10,n)
X,Y = np.meshgrid(x,y)
contourf(X, Y, f(X,Y), 8, alpha=0, cmap=cm.hot)
C = contour(X, Y, f(X,Y), 8, colors='black', linewidth=0.5)
clabel(C, inline=1, fontsize=10)
show()
```

```python
from matplotlib.pyplot import *
def f(x,y):
    return (1-x)**2 +(1-y)**2

n = 256
x = np.linspace(-100,100,n)
y = np.linspace(-100,100,n)
X,Y = np.meshgrid(x,y)
contourf(X, Y, f(X,Y), 20, alpha=0.75, cmap=cm.hot)#alpha=0.25表示颜色 20线的个数
C = contour(X, Y, f(X,Y), 21, colors='black', linewidth=0.01)# 21黑线的个数
clabel(C, inline=0.1, fontsize=10)
show()
```
